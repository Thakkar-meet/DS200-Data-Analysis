```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("mpg")
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
colors = {3:'red',4:'orange',5:'green',6:'blue',8:'purple'}

plt.figure(figsize=(8,6))
for c in df["cylinders"].unique():
    subset = df[df["cylinders"] == c]
    plt.scatter(subset["weight"], subset["mpg"], alpha=0.6, label=f'{c} cylinders', color=colors[c])
plt.title("Weight vs MPG (colored by Cylinders)")
plt.xlabel("Weight")
plt.ylabel("Miles per Gallon (MPG)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
df.boxplot(column="mpg", by="cylinders", grid=False)
plt.title("Fuel Efficiency Distribution by Number of Cylinders")
plt.suptitle("")
plt.xlabel("Cylinders")
plt.ylabel("Miles per Gallon (MPG)")
plt.show()
```

```python
avg_mpg_by_year = df.groupby("model_year")["mpg"].mean()

plt.figure(figsize=(8,6))
plt.plot(avg_mpg_by_year.index, avg_mpg_by_year.values, marker='o', color='green')
plt.title("Average MPG Over the Years")
plt.xlabel("Model Year")
plt.ylabel("Average Miles per Gallon (MPG)")
plt.grid(True)
plt.show()
```

```python
avg_mpg_by_origin = df.groupby("origin")["mpg"].mean()

plt.figure(figsize=(7,5))
plt.bar(avg_mpg_by_origin.index, avg_mpg_by_origin.values, color=['skyblue', 'orange', 'lightgreen'])
plt.title("Average Fuel Efficiency by Origin")
plt.xlabel("Origin")
plt.ylabel("Average MPG")
plt.show()
```
